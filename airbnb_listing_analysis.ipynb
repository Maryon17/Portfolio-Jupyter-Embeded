# **AirBnB Listing Analysis**
The goal of this project is to analyze AirBnB listings in Paris to assess the impact of recent short-term rental regulations. By cleaning, preparing, and visualizing the data, the analysis aims to uncover trends in host activity, pricing patterns, and neighborhood-level differences before and after the regulatory changes. The insights generated will help identify how the Paris rental market has evolved in response to these policies.

### Objective 1
## **Profile & QA the data**
**Tasks:**
* Read in the AirBNB listing data.
* Cast any date columns as a datetime format.
* Filter down the data to just listing of the city of Paris.
* QA the paris listing: check for missing values, and calcuate the minimum and maximum, and average for each numeric field
  


```python
import pandas as pd

listings = pd.read_csv(
    "Listings.csv",
    encoding="ISO-8859-1",
    low_memory=False,
    parse_dates=["host_since"]
)
```

```python
listings.head()
```

```python
listings.info()
```

```python
paris_listings = (
    listings
    .query("city == 'Paris'")
    .loc[:,["host_since", "neighbourhood", "city", "accommodates", "price"]]
)

paris_listings.info()
```

```python
paris_listings.describe()
```

### Objective 2
## **Prepare the data for visualization**
**Tasks:**
* Create a DataFrame called **paris_listings_neighbourhood**. Group the Paris listings by neighbourhood and calculate the average price for each. Sort by price in ascending order.
* Create a DataFrame called **paris_listings_accommodations**. Filter data down to the most expensive neighbourhood in Paris. Group it by accomodates and calculate the average price for each. Sort by price in ascending order.
* Create a DataFrame called **paris_listings_over_time**. Group the data by the year compopnent of host_since. Calculate a count of rows to get the number of new hosts of each year, and the average price of listings for each year.

```python
paris_listings_neighbourhood = (
    paris_listings.
    groupby("neighbourhood")
    .agg({"price":"mean"})
    .sort_values("price")
)

paris_listings_neighbourhood.tail()
```

```python
paris_listings_accommodates =(
    paris_listings
    .query("neighbourhood == 'Elysee'")
    .groupby("accommodates")
    .agg({"price":"mean"})
    .sort_values("price")
)

paris_listings_accommodates.tail()
```

```python
paris_listings_over_time =(
    paris_listings
    .set_index("host_since")
    .resample("YE")
    .agg({
        "neighbourhood":"count",
        "price":"mean",
    })
)

paris_listings_over_time.tail()
```

### Objective 3
## **Visualize the data**
**Tasks:**
* Build a horizontal bar chart of average rent price by neighborhood.
* Build a horizontal bar chart of average price by accomodates in the most expensive neighbourhood
* Build a line chart of new hosts per year and average price by year

**Questions to answer:**

```python
import seaborn as sns

(paris_listings_neighbourhood
 .plot
 .barh(
     title="Average Listing Price by Paris Neighbourhood",
     xlabel="Price Per Night (Euros)",
     ylabel="Neighbourhood",
     legend=None
 )
)

sns.despine()
```

```python
(paris_listings_accommodates
 .plot
 .barh(
     title="Average Listing Price by Accomodation Number",
     xlabel="Price Per Night (Euros)",
     ylabel="Accommodation Capacity",
     legend=None
 )
)

sns.despine()
```

```python
paris_listings_over_time["neighbourhood"].plot(
    title="New Airbnb Host in Paris Over Time",
    ylabel="New Hosts"
)

sns.despine()
```

```python
paris_listings_over_time["price"].plot(
    title="Average AirBnB Price in Paris Over Time",
    ylabel="Average Price (Euros)"
)

sns.despine()
```

```python
import matplotlib.pyplot as plt

fig, ax =plt.subplots()

ax.plot(
    paris_listings_over_time.index,
    paris_listings_over_time["neighbourhood"],
    label="New Hosts",
    c="pink"
)

ax.set_ylabel("New Hosts")

ax2 = ax.twinx()

ax2.plot(
    paris_listings_over_time.index,
    paris_listings_over_time["price"],
    label="Average Price",
)

ax2.set_ylim(0)

ax2.set_ylabel("Average Price")

ax.set_title("2015 Regulations Lead to Fewer New Hosts, Higher Prices")
```

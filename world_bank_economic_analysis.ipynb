# World Bank Economic Analysis

The goal of this project is to analyze the World Bank's annual economic report and produce an infographic-style report highlighting major trends in global economic growth and development.

### Objective 1
## **Profile & QA the data**
**Tasks:**
* Import the World Bank Excel file
* Use "GDP" and "GDP per Capita" to calculate population in millions and create a "Population (M)" column.
* Perform basic profiling & QA on the data.
* Filter the data to 2014 and join it with the HDI csv file on "Country Code".
  


```python
import pandas as pd
import numpy as py
```

```python
world_bank = pd.read_excel("../WorldBank.xlsx")
world_bank["Population (M)"] = (world_bank["GDP (USD)"]/world_bank["GDP per capita (USD)"]) / 1_000_000
world_bank = world_bank.rename(columns={'Country Name': 'Country'})

world_bank.head()
```

```python
world_bank.info()
```

```python
world_bank.describe()
```

```python
hdi = pd.read_csv("../HDI.csv")

hdi.head()
```

```python
wb_hdi_2014 = (world_bank
               .query("Year == 2014")
               .merge(
                   hdi[["iso3","hdi_2014"]],
                   how="left",
                   left_on="Country Code",
                   right_on="iso3"
               ).drop("iso3",axis=1)
              )
wb_hdi_2014.head()
```

### Objective 2
## **Prepare the data for visualization**
**Tasks:**
* Create a "gdp_pivot" table with years as rows, regions as columns, sum of GDP as values.
* Create "pop_pivot" table with years as rows, regions as columns, and sum of population as values.
* Create "wb_hdi_by_region" table by calculating the average HDI for each region and sorting from highest average HDI to lowest
  


```python
gdp_pivot = (world_bank
             .pivot_table(
                 index=world_bank["Year"],
                 columns=world_bank["Region"],
                 values="GDP (USD)",
                 aggfunc="sum"
             )
            )
    
gdp_pivot.head()         
```

```python
pop_pivot = (world_bank
             .pivot_table(
                 index=world_bank["Year"],
                 columns=world_bank["Region"],
                 values="Population (M)",
                 aggfunc="sum"
             )
            )
    
pop_pivot.head()         
```

```python
wb_hdi_by_region = (wb_hdi_2014
                    .groupby("Region")
                    .agg({"hdi_2014":"mean"})
                    .sort_values("hdi_2014", ascending=False)
                   )

wb_hdi_by_region.head()
```

### Objective 3
## **Visualize the data**
**Tasks:**
* Create a stacked area chart showing the growth of GDP over time, with each stack representing a geographic region.
* Create a stacked area chart showing the growth of population over time, with each stack representing a geographic region.
* Create a bubble chart with life expectancy for each country in the x-axis, GDP per capita in the y-axis (log scale), and population as the bubble size. Color the bubbles by region using the same colors for each as in the previous charts.
* Create a bar chart to show average HDI by region (make sure the region colors match)
* Create a scatterplot with power consumption in the x-axis and GDP per capita in the y-axis, then use HDI as a color scale for the dots (make sure to remove outliers like Iceland).
  


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gdp_pivot.iloc[-2:-1].T
```

```python
fig, ax = plt.subplots()

ax.stackplot(
    gdp_pivot.index,
    [gdp_pivot[region] / 1_000_000_000_000 for region in gdp_pivot.iloc[-1].sort_values(ascending=False).index],
    labels=gdp_pivot.iloc[-1].sort_values(ascending=False).index
)

ax.legend()
ax.set_title("GDP as Grown Exponentially Over Time")
ax.set_ylabel("GDP (Trillions)")


plt.show()
```

```python
fig, ax = plt.subplots()

ax.stackplot(
    pop_pivot.index,
    [pop_pivot[region] / 1000 for region in pop_pivot.iloc[-1].sort_values(ascending=False).index],
    labels= pop_pivot.iloc[-1].sort_values(ascending=False).index
)

ax.legend()
ax.set_title("Population has surged from 2 billion to 7.5 billion")
ax.set_ylabel("Population (Billions)")
plt.show()
```

```python
import matplotlib.ticker as mticker

fig, ax = plt.subplots()

minsize= min(wb_hdi_2014["Population (M)"])
maxsize= max(wb_hdi_2014["Population (M)"])


sns.scatterplot(
    x="Life expectancy at birth (years)",
    y="GDP per capita (USD)",
    data=wb_hdi_2014,
    size="Population (M)",
    sizes=(minsize,maxsize),
    hue="Region",
    ax=ax
).set(title="Life Expectance Increases as Countires get richer",yscale="log")  

ax.yaxis.set_major_formatter(mticker.ScalarFormatter())
```

```python
wb_hdi_by_region.plot.bar(title = "HDI by Region", ylabel="Human Development Indicator")
```

```python
sns.scatterplot(
    data=wb_hdi_2014.query("Country != 'Iceland'"),
    x="Electric power consumption (kWh per capita)",
    y="GDP per capita (USD)",
    hue="hdi_2014",
    palette="coolwarm_r"
).set(title="Electricity Drive Development")
```

### Objective 4
## **Build a final report**
**Tasks:**
* Assemble the 5 visualizations you have created into a single page.
* Add an overall title and descriptive text to add context for the reader.
  


```python
import matplotlib.gridspec as gridspec
import matplotlib.collections as mcoll
import matplotlib as mpl

fig = plt.figure(figsize=(8,12))

gs = gridspec.GridSpec(nrows=12, ncols=6)

stack_list = ['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3', '#a6d854', '#ffd92f', '#b3b3b3']
stack_list2 = ['#66c2a5', '#a6d854', '#b3b3b3', '#fc8d62', '#e78ac3', '#8da0cb', '#ffd92f']
bar_list = ['#8da0cb', '#fc8d62', '#e78ac3', '#ffd92f', '#66c2a5', '#a6d854', '#b3b3b3']
bubble_list = ['#a6d854', '#fc8d62', '#ffd92f', '#66c2a5', '#b3b3b3', '#e78ac3', '#8da0cb']

fig.suptitle(
    "Report: Global Economic Development from 1960-2018",
    fontsize=16,
    fontweight="bold",
    x=.65,
    y=.93
)

ax1 = fig.add_subplot(gs[0:4, 0:3])

ax1.stackplot(
    gdp_pivot.index,
    [gdp_pivot[region] / 1_000_000_000_000 for region in gdp_pivot.iloc[-1].sort_values(ascending=False).index],
    labels=gdp_pivot.iloc[-1].sort_values(ascending=False).index,
    colors=stack_list
)

ax1.set_title("GDP as Grown Exponentially Over Time")
ax1.set_ylabel("GDP (Trillions)")

fig.text(.92,.61,
'''
Economic Expansion in the 20th and early
21st Centuries has been enourmous and has
outpaced population growth. In 1960, Europe
and North America produced 75% of GDP on
20% of Population.

By 2018, East Asia had surpassed both
in economic output largely due to the
rise in Japan, South Korea and China.

In the coming decades, economists expect
South Asia and Sub-Saharan African to lean
the world in global growth
'''
,fontsize=8)

ax2 = fig.add_subplot(gs[0:4, 3:6])

ax2.stackplot(
    pop_pivot.index,
    [pop_pivot[region] / 1000 for region in pop_pivot.iloc[-1].sort_values(ascending=False).index],
    labels= pop_pivot.iloc[-1].sort_values(ascending=False).index,
    colors=stack_list2
)

ax2.legend(bbox_to_anchor=(1.75,1),fontsize=8)
ax2.set_title("Population has surged from 2 billion to 7.5 billion")
ax2.set_ylabel("Population (Billions)")


ax3 = fig.add_subplot(gs[4:8, :])

minsize= min(wb_hdi_2014["Population (M)"])
maxsize= max(wb_hdi_2014["Population (M)"])


sns.scatterplot(
    x="Life expectancy at birth (years)",
    y="GDP per capita (USD)",
    data=wb_hdi_2014,
    size="Population (M)",
    sizes=(minsize,maxsize),
    hue="Region",
    palette=bubble_list,
    ax=ax3
).set(title="Life Expectance Increases as Countires get richer",yscale="log")

ax3.yaxis.set_major_formatter(mticker.ScalarFormatter())

handles, labels = ax3.get_legend_handles_labels()
entries_to_skip = len(wb_hdi_2014["Region"].unique())+1

ax3.legend(
    handles[entries_to_skip:],
    labels[entries_to_skip:],
    bbox_to_anchor=(.8, .9),
    borderaxespad=0,
    frameon=False,
    ncol=6,
    fontsize=8
)

fig.text(.92, .45,
'''
The wealthy nations of the world
enjoy high GDP per capita as well
as long life spans, but make up
relatively small share of the
global population.

If growth can continue in the developing
world, humanita will be vastly wealthier,
happier, and (hopefully) happier.
'''
,fontsize=8)
                      
ax4 = fig.add_subplot(gs[8:12, 0:3])

ax4.bar(
    wb_hdi_by_region.index,
    wb_hdi_by_region.values.flatten(),
    color=bar_list
)
ax4.set_title("HDI by Region")
ax4.set_ylabel("Human Development Indicator")
ax4.tick_params(axis='x', rotation=90)

ax5 = fig.add_subplot(gs[8:12, 3:6])

sns.scatterplot(
    data=wb_hdi_2014.query("Country != 'Iceland'"),
    x="Electric power consumption (kWh per capita)",
    y="GDP per capita (USD)",
    hue="hdi_2014",
    palette="coolwarm_r",
    ax=ax5
).set(title="Electricity Drive Development")

fig.text(.92, .13,
'''
HDI, short for Human Development Index,
attempts to measure the overall standard
of living in countries. Life Expectancy,
GDP per Capita, and Educational attainment
are the factors that are considered.

Economic growth in the developing world
in the 21st century should help other regions
catch up to North America and Europe.

One factor not in HDI, but instrumental in
a country's development, is electricty consumption.
Electricity unlocks massive improvements in
productivity. Developing nations should continue
to invest in energy production to ensure growth.
'''
,fontsize=8)

fig.subplots_adjust(wspace=4, hspace=4)
```
